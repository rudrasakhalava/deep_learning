```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_json("students.json")
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df["student_iq"],y=df["student_cgpa"],hue=df["placement"])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()

p.fit(X,y)
```

```python
print(p.coef_)
print(p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values,y.values,clf=p,legend=2)
```
